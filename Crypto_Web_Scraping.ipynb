```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
import re
```

```python
# Requesting to scrap Data from web
url='https://markets.businessinsider.com/cryptocurrencies'
req=requests.get(url)
```

```python
# Checking status code
req
```

```python
# Scraping Web Data
soup=BeautifulSoup(req.content)
```

```python
arr1=[]
for i in soup.findAll('td'):
  arr1.append(i)
```

```python
# Initialising a Dictionary to store all the scrapped Data
crypto={'Name':[],'Price':[],'Market_cap':[],'Circulation':[],'Volume':[]}

# To seperate the values and append them in the correct column
count=0
for i in arr1:

  if count==0:
    count+=1
    crypto['Name'].append(re.sub('^<tr.*r">|<td.*r">|</a></td>|\n|<td.*>|</tr>|^<.*>|<.*>','',str(i)))

  elif count==1:
    count+=1
    crypto['Price'].append(re.sub('^<td .*">|</span> </div></td>|\n|<.*>','',str(i)))

  elif count==2:
    count+=1
    # Continue is used since we are taking that data into account
    continue

  elif count==3:
    count+=1
    continue

  elif count==4:
    count+=1
    crypto['Market_cap'].append(re.sub('<td.*">|</td>|\n.*\n','',str(i)))

  elif count==5:
    count+=1
    crypto['Circulation'].append(re.sub('<td.*">|</td>|\n.*\n','',str(i)))

  elif count==6:
    crypto['Volume'].append(re.sub('<td.*">|</td>|\n.*\n','',str(i)))
    count+=1

  elif count==7:
    count=0
    continue
```

```python
# Equalising the length of the Columns
if not len(crypto['Name'])==len(crypto['Price'])==len(crypto['Market_cap'])==len(crypto['Circulation'])==len(crypto['Volume']):
  crypto['Volume'].append(np.nan)
```

```python
# Creating DataFrame
crypto_data=pd.DataFrame(crypto,columns=crypto.keys())
```

```python
# Removing unwanted Data
crypto_data=crypto_data[crypto_data.Name != '']
```
